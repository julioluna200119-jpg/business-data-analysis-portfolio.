```python
!pip install ipython-sql

%%sql

-- PROYECTO: PERFORMANCE FINANCIERO ADVENTURE WORKS
-- Objetivo: Analizar rentabilidad por país y retorno de inversión (ROI)

-- 1. Unión de tablas y limpieza de datos (Sales + Products + Territory)
SELECT
    v.numero_pedido,
    p.nombre_producto,
    pc.clave_categoria,
    COALESCE(p.precio_producto, 0) AS precio_unitario,
    COALESCE(v.cantidad_pedido, 0) AS cantidad,
    COALESCE(p.costo_producto, 0) AS costo_unitario,
    t.pais
FROM ventas_2017 AS v
LEFT JOIN productos AS p ON v.clave_producto = p.clave_producto
LEFT JOIN productos_categorias AS pc ON p.clave_subcategoria = pc.clave_subcategoria
LEFT JOIN territorios AS t ON v.clave_territorio = t.clave_territorio;

-- 2. Cálculo de KPIs Financieros por País (Costo vs Ingreso vs ROI)
-- Hallazgo: USA tiene el mejor ROI con 75.7%
SELECT
    p.pais,
    SUM(p.ingresos)::integer AS total_revenue,
    SUM(p.costos)::integer AS total_cost,
    SUM(p.ingresos)::integer - SUM(p.costos)::integer AS gross_profit,
    ROUND((SUM(p.ingresos) - SUM(p.costos))*100.0 / NULLIF(SUM(p.ingresos), 0), 2) AS margin_pct,
    ROUND((SUM(p.ingresos) - SUM(p.costos))*100.0 / NULLIF(SUM(c.costo_campana), 0), 2) AS ROI_pct
FROM pais_ingreso_costo AS p
LEFT JOIN pais_campanas AS c ON p.clave_territorio = c.clave_territorio
GROUP BY p.pais
ORDER BY total_revenue DESC;
```
